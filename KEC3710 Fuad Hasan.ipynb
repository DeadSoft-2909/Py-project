```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np

# Load the dataset
file_path = '/content/drive/MyDrive/Depression Student Dataset.csv'
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    exit()

# Dataset details
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)

# Descriptive statistics
print("\nDescriptive Statistics:\n", df.describe(include='all'))

# Duplicate values
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)

# Handle missing values in 'Clean/Update/Mobile' column (replace with the mode)
# Get the actual column name (case-sensitive)
# Check if any column contains 'Clean' before accessing by index
clean_columns = df.columns[df.columns.str.contains('Clean', case=False)]
if len(clean_columns) > 0:
    column_name = clean_columns[0]
    # Use the actual column name to calculate the mode
    mode_mobile = df[column_name].mode()[0]
    df[column_name] = df[column_name].fillna(mode_mobile)

    # Verify if missing values are filled
    print("\nMissing values after handling 'Clean/Update/Mobile':\n", df[column_name].isnull().sum())
else:
    print("Warning: No column found containing 'Clean'. Skipping missing value handling for this column.")


# Calculate mean, median, mode, and standard deviation for numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
for col in numerical_cols:
    print(f"\nStatistics for {col}:")
    print(f"  Mean: {df[col].mean()}")
    print(f"  Median: {df[col].median()}")
    print(f"  Mode: {df[col].mode()[0]}") # Taking the first mode if multiple exist
    print(f"  Standard Deviation: {df[col].std()}")
```
